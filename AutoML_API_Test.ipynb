```python
import requests, pandas as pd, json
```

```python
credit_data_original = pd.read_csv('credit_data.csv').sample(10).reset_index(drop=True)
credit_data_original
```

```python
credit_data = credit_data_original.drop(columns=['customer_type'],inplace=False)
credit_data
```

```python
testjson = credit_data.to_json(orient="records")
testjson
```

```python
credit_data_apiurl = "http://localhost:5000/automl_api/v1/Credit_Score_Model"
```

```python
data = {"testjson":json.dumps(testjson)}
out = requests.post(credit_data_apiurl, data)
```

```python
out = json.loads(out.text)
out
```

```python
out = pd.read_json(out['data'])
```

```python
for col in out.columns:
    credit_data_original[col] = out[col]
credit_data_original
```
